# reading data

# filtering data

We only want data for summer olympics=>

```python
df=df[df['Season']=='Summer']
df.head()
```

## Get name of country from *NOC* using the NOC_DATA by merging it

```python
df=df.merge(region_df,on='NOC',how='left')
df.head()
```

## Removing duplicated rows

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Medal'].value_counts()
```

```python
pd.get_dummies(df['Medal'])
```

We concatenate the above dataframe with the main df

We will horizontally stack them

```python
df=pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
```

```python
df.head()
```

## now we will find the sum of medals by NOC, by groupby of all columns

```python
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False)
```

```python
#reset index
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

**NOTE**: There is an issue, rather than counting single medal of one game, in this above data medals of each members in a team is being counted. To fix this issue will remove the duplicated rows and then count the medals(acc to team, noc, year,season,city)

```python
medal_tally=df.drop_duplicates(subset=['NOC','Sport','Team','Year','Medal','City','Event'])
medal_tally.shape
```

```python
medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```
